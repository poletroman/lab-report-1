```python
import statistics

data = [1.09, 1.53, -3.52, -0.29, -1.10, -1.88]

std_deviation = statistics.stdev(data)

print(f"The standard deviation of the numbers is: {std_deviation:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

acceleration = [4.9, -6.3, 2.4, 0.2, -4.0, 6.3]
force = [4.46, -4.11, 0.68, 0.67, -3.63, 3.35]


std_dev = 1.89

fig, ax = plt.subplots()

ax.errorbar(acceleration, force, xerr=std_dev, fmt='o', label='Data with Error Bars')

slope, intercept = np.polyfit(acceleration, force, 1)

x_fit = np.linspace(min(acceleration) - 2 * std_dev, max(acceleration) + 2 * std_dev, 100)
y_fit = slope * x_fit + intercept


ax.plot(x_fit, y_fit, 'r', label=f'Best Fit (slope = {slope:.2f})')

ax.set_xlabel('Acceleration')
ax.set_ylabel('Force')
ax.legend()

plt.title("Acceleration vs Force")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


a_values = [0.229, 0.584, 0.855, 0.516, 0.670, 0.825]
mass_values = [0.023, 0.060, 0.087, 0.053, 0.068, 0.084]

fig, ax = plt.subplots()

ax.scatter(mass_values, a_values, label='Data Points')


slope, intercept = np.polyfit(mass_values, a_values, 1)


x_fit = np.linspace(min(mass_values), max(mass_values), 100)
y_fit = slope * x_fit + intercept

ax.plot(x_fit, y_fit, 'r', label=f'Best Fit (slope = {slope:.3f})')


ax.set_xlabel('M* (kg)')
ax.set_ylabel('a')
ax.legend()


plt.title('a vs M* with Best Fit Line')
plt.grid(True)
plt.show()


```

```python
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt


def calculate_angular_acceleration(df, time_column, angular_velocity_column):

  slope, intercept, r_value, p_value, std_err = linregress(df[time_column], df[angular_velocity_column])


    angular_acceleration = slope


    uncertainty = std_err

    return angular_acceleration, uncertainty

def main():

    file_paths = ['Capstone Data 1.csv', 'Capstone Data 2.csv', 'Capstone Data 3.csv', 'Capstone Data (3).csv']

    for file_path in file_paths:

        df = pd.read_csv(file_path)


        time_column = 'Time (s) Run #1'
        angular_velocity_column = 'Angular Velocity (rad/s) Run #1'

        angular_acceleration, uncertainty = calculate_angular_acceleration(df, time_column, angular_velocity_column)


        print(f"File: {file_path}")
        print(f"Angular Acceleration: {angular_acceleration} rad/s^2")
        print(f"Uncertainty: {uncertainty} rad/s^2")
        print()


        plt.scatter(df[time_column], df[angular_velocity_column], label='Data')
        plt.plot(df[time_column], angular_acceleration * df[time_column], color='red', label='Fitted Line')
        plt.xlabel('Time (s)')
        plt.ylabel('Angular Velocity (rad/s)')
        plt.legend()
        plt.show()

if __name__ == "__main__":
    main()
```
